# Spotify Popularity Data 2024 - An Exploratory Data Analysis

This is a Spotify Popularity Data EDA using this dataset (https://www.kaggle.com/datasets/nelgiriyewithana/most-streamed-spotify-songs-2024/data). In this code, we're about to explore the most popular songs and artists through Spotify streams, the trend of Tiktok, YouTube, Soundcloud, and Spotify streams throughout the music release years, as well as analyzing the musical trends of two of the most popular artists in the world: Taylor Swift and The Weeknd.

# Data Preparation

```python
# Import the libraries that will be used
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Most Streamed Spotify Songs 2024.csv', index_col=0, encoding='ISO-8859-1')
```

```python
df1.head(10)
```

# Data Cleaning

## Duplicate Checking

```python
print(f"Total of rows in our data: {df1.shape[0]}")
print(f"Total of duplicated rows: {df1.duplicated().sum()}")
```

```python
df1 = df1.drop_duplicates()
```

```python
print(f"Total of rows in our data after erasing duplicates: {df1.shape[0]}")
print(f"Total of duplicated rows: {df1.duplicated().sum()}")
```

## Dealing with Missing Values

```python
df1.info()
```

```python
# Count the percentage of missing values in each column
(df1.isna().sum() / df1.shape[0]) * 100
```

```python
# Drop TIDAL Popularity Column since it has no values whatsoever.
df1 = df1.drop('TIDAL Popularity', axis=1)
```

```python
# Drop rows of data that doesn't have an Artist name.
df1 = df1.dropna(subset=['Artist'])
df1 = df1[df1['Artist'].str.strip() != '']
```

```python
df1.select_dtypes(include=['object']).columns
```

```python
# Convert the columns below to float
columns_to_convert = [
    'All Time Rank', 'Spotify Streams', 'Spotify Playlist Count', 'Spotify Playlist Reach',
    'YouTube Views', 'YouTube Likes', 'TikTok Posts', 'TikTok Likes', 'TikTok Views',
    'YouTube Playlist Reach', 'AirPlay Spins', 'SiriusXM Spins', 'Deezer Playlist Reach',
    'Pandora Streams', 'Pandora Track Stations', 'Soundcloud Streams', 'Shazam Counts'
]

for column in columns_to_convert:
    df1[column] = df1[column].str.replace(',', '').astype(float)
```

```python
# Fill in the columns below with median
columns_to_fill = [
    'Spotify Streams', 'Spotify Playlist Count', 'Spotify Playlist Reach',
    'Spotify Popularity', 'YouTube Views', 'YouTube Likes', 'TikTok Posts',
    'TikTok Likes', 'TikTok Views', 'YouTube Playlist Reach',
    'Apple Music Playlist Count', 'AirPlay Spins', 'SiriusXM Spins',
    'Deezer Playlist Count', 'Deezer Playlist Reach', 'Amazon Playlist Count',
    'Pandora Streams', 'Pandora Track Stations', 'Soundcloud Streams', 'Shazam Counts'
]

for column in columns_to_fill:
    median_value = df1[column].median()
    df1[column] = df1[column].fillna(median_value)
```

```python
(df1.isna().sum() / df1.shape[0]) * 100
```

```python
for column in columns_to_fill:
    df1[column] = df1[column].astype(int)

df1["All Time Rank"] = df1["All Time Rank"].astype(int)
df1["Track Score"] = df1["Track Score"].astype(int)
```

```python
# Convert Release Date into Datetime Object Type
df1['Release Date'] = pd.to_datetime(df1['Release Date'], format="%m/%d/%Y")
```

```python
df1['Release Date'].apply(type).unique()
```

```python
df1.to_csv('cleaned_spotify_2024_data.csv', index=False)
```

# Data Analysis

First, we create the formatters to clean up the axes' scale.

```python
from matplotlib.ticker import FuncFormatter

def billions(x, pos):
    return f'{x*1e-9:.1f}'

billion_formatter = FuncFormatter(billions)

def millions(x, pos):
  return f'{x*1e-6:.1f}'

million_formatter = FuncFormatter(millions)

def thousands(x, pos):
  return f'{x*1e-3:.1f}'

thousand_formatter = FuncFormatter(thousands)
```

We create a separate column for the release year of the music.

```python
df1['Release Year'] = df1['Release Date'].dt.year
```

## Analyzing Top 10 Biggest Songs of 2024 Through Spotify Streams

```python
from matplotlib.ticker import FuncFormatter

all_time_ranks = df1.head(10)

# Plot the data
plt.figure(figsize=(8, 6))
plt.barh(all_time_ranks.index, all_time_ranks['Spotify Streams'], color='rebeccapurple')
plt.xlabel('Spotify Streams (Billions)')
plt.ylabel('Songs')
plt.title('Spotify Streams of The Biggest Songs in 2024 (in billion)')
plt.gca().invert_yaxis()
plt.grid(True)
plt.gca().xaxis.set_major_formatter(billion_formatter)
plt.show()
```

## Analyzing Top 10 Most Popular Artists Through Spotify Streams

```python
most_popular_artists = df1.groupby('Artist')['Spotify Streams'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 8))
most_popular_artists.plot(kind='barh', color='rebeccapurple')
plt.title('Most Popular Artists Through Spotify Streams: 2024 (in Billions)')
plt.xlabel('Spotify Streams (Billions)')
plt.ylabel('Artist')
plt.gca().invert_yaxis()
plt.gca().xaxis.set_major_formatter(billion_formatter)
plt.show()
```

## Analyzing Top 10 Most Trending Artists Through Spotify Streams

```python
most_trending_artists = df1.groupby('Artist')['Spotify Popularity'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 8))
most_trending_artists.plot(kind='barh', color='rebeccapurple')
plt.title('Most Trending Artists Through Spotify Popularity: 2024 (in Billions)')
plt.xlabel('Spotify Popularity')
plt.ylabel('Artist')
plt.gca().invert_yaxis()
plt.show()
```

## Analyzing The Sum of TikTok Views Throughout Music Release Years

```python
tiktok_views_per_year = df1.groupby('Release Year')['TikTok Views'].sum()

plt.figure(figsize=(10, 5))
tiktok_views_per_year.plot(kind='line', color='darkslateblue', marker='o')
plt.title('The Sum of Tiktok Views per Year (in Billion)')
plt.xlabel('Music Release Year')
plt.ylabel('TikTok Views (Billions)')
plt.grid(True)
plt.gca().yaxis.set_major_formatter(billion_formatter)
plt.show()

```

There are spikes on the year 2005 and 2011, let's analyze what song(s) that causes these spikes on each respective year.

```python
# Look for the most viewed tiktok song of 2005
filtered_df = df1[df1["Release Year"] == 2005]

filtered_df = filtered_df[["Artist", "TikTok Views"]]

filtered_df.sort_values(by="TikTok Views", ascending=False).head(3)
```

```python
# Look for the most viewed tiktok song of 2011

filtered_df = df1[df1["Release Year"] == 2011]

filtered_df = filtered_df[["Artist", "TikTok Views"]]

filtered_df.sort_values(by="TikTok Views", ascending=False).head(3)
```

## Analyzing The Sum of TikTok Posts Throughout Music Release Years

```python
tiktok_posts_per_year = df1.groupby('Release Year')['TikTok Posts'].sum()

plt.figure(figsize=(10, 5))
tiktok_posts_per_year.plot(kind='line', color='darkslateblue', marker='o')
plt.title('The Sum of Tiktok Posts per Year (in Million)')
plt.xlabel('Music Release Year')
plt.ylabel('TikTok Posts (Millions)')
plt.grid(True)
plt.gca().yaxis.set_major_formatter(million_formatter)
plt.show()
```

## Analyzing The Sum of TikTok Likes Throughout Music Release Years

```python
tiktok_likes_per_year = df1.groupby('Release Year')['TikTok Likes'].sum()

plt.figure(figsize=(10, 5))
tiktok_likes_per_year.plot(kind='line', color='darkslateblue', marker='o')
plt.title('The Sum of Tiktok Likes per Year (in Billions)')
plt.xlabel('Music Release Year')
plt.ylabel('TikTok Likes (Billions)')
plt.grid(True)
plt.gca().yaxis.set_major_formatter(billion_formatter)
plt.show()
```

From the three graphs above, we can conclude that music released from 2015 onwards has an drastic gap in popularity in Tiktok, and it has a significant rise on 2020 where TikTok songs truly blew up during the heights of pandemic.

## Analyzing The Sum of SoundCloud Streams Throughout Music Release Years

```python
soundcloud_streams_per_year = df1.groupby('Release Year')['Soundcloud Streams'].sum()

plt.figure(figsize=(10, 5))
soundcloud_streams_per_year.plot(kind='line', color='darkorange', marker='o')
plt.title('The Sum of Soundcloud Streams per Year (in Billions)')
plt.xlabel('Music Release Year')
plt.ylabel('Soundcloud Streams (Billions)')
plt.grid(True)
plt.gca().yaxis.set_major_formatter(billion_formatter)
plt.show()
```

From the three graphs above, we can conclude that music released from during the later stages of the 2010s reaches its first peak of popularity, before declining and reaching its peak again in 2023.

## Analyzing YouTube Views and Spotify Streams Throughout Music Release Years

```python
spotify_popularity = df1.groupby('Release Year')['Spotify Streams'].sum()
youtube_views_per_year = df1.groupby('Release Year')['YouTube Views'].sum()

plt.figure(figsize=(10, 5))
plt.plot(youtube_views_per_year.index, youtube_views_per_year.values, label='YouTube Views', color='red', marker='o')
plt.plot(spotify_popularity.index, spotify_popularity.values, label='Spotify Streams', color='darkgreen', marker='o')

plt.title('YouTube Views and Spotify Streams per Year (in Billions)')
plt.xlabel('Music Release Year')
plt.ylabel('Views/Streams (Billions)')
plt.gca().yaxis.set_major_formatter(billion_formatter)
plt.grid(True)
plt.legend()
plt.show()
```

From the line chart, we can conclude that YouTube reached its peak of popularity of music views in 2017, before declining. Meanwhile, Spotify became the main streaming service during the later stages of 2010s, and continued to rise in the 2020s.

## Analyzing Taylor Swift's Spotify Streams Throughout Music Release Years

```python
taylor_swift = df1[df1["Artist"] == "Taylor Swift"]
taylor_swift_popularity = taylor_swift.groupby('Release Year')['Spotify Streams'].sum()

plt.figure(figsize=(10, 5))
taylor_swift_popularity.plot(kind='line', marker='o', color='darkgreen')
plt.title("Taylor Swift's Spotify Streams (in billions) Throughout the Music Release Years")
plt.xlabel('Music Release Year')
plt.ylabel('Spotify Streams (billions)')
plt.gca().yaxis.set_major_formatter(billion_formatter)
plt.grid(True)
plt.show()
```

## Analyzing Taylor Swift's Most Popular Albums through Spotify Streams

```python
most_popular_albums_taylor_swift = taylor_swift.groupby('Album Name')['Spotify Streams'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8, 6))
most_popular_albums_taylor_swift.plot(kind='bar', color='darkgreen')
plt.title('Most Popular Taylor Swift Albums through Spotify Streams (in Billions)')
plt.xlabel('Album Name')
plt.ylabel('Spotify Streams (billions)')
plt.gca().yaxis.set_major_formatter(billion_formatter)
plt.show()
```

We can conclude that Taylor's music always rises whenever she drops three of her most popular albums. "1989" was released in 2014, "Midnights" was released in 2022, "Lover" was released in 2019. On those years, her Spotify streams increase.

## Analyzing The Weeknd's Spotify Streams Throughout Music Release Years

```python
the_weeknd = df1[df1["Artist"] == "The Weeknd"]
the_weeknd_popularity = the_weeknd.groupby('Release Year')['Spotify Streams'].sum()

plt.figure(figsize=(10, 5))
the_weeknd_popularity.plot(kind='line', marker='o', color='darkgreen')
plt.title("The Weeknd's Spotify Streams (billions) Throughout the Music Release Years")
plt.xlabel('Music Release Year')
plt.ylabel('Spotify Streams (billions)')
plt.grid(True)
plt.gca().yaxis.set_major_formatter(billion_formatter)
plt.show()
```

## Analyzing The Weeknd's Most Popular Albums through Spotify Streams

```python
most_popular_albums_the_weeknd = the_weeknd.groupby('Album Name')['Spotify Streams'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8, 6))
most_popular_albums_the_weeknd.plot(kind='bar', color='darkgreen')
plt.title('Most Popular The Weeknd Albums through Spotify Streams (in Billions)')
plt.xlabel('Album Name')
plt.ylabel('Spotify Streams (billions)')
plt.gca().yaxis.set_major_formatter(billion_formatter)
plt.show()
```

We can conclude that "Starboy" and "After Hours" (which contains "Blinding Lights") are two of The Weeknd's most popular albums, and two of his Spotify streams reach their peak in those albums' release years (2016 and 2020).
