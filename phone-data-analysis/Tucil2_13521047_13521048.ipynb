Exploratory Data Analysis (EDA) adalah pendekatan statistik untuk menganalisis dataset dengan tujuan untuk memahami struktur dan karakteristik data, serta mengidentifikasi pola, hubungan, anomali, dan kecenderungan yang mungkin terdapat dalam data. EDA membantu dalam merumuskan hipotesis, mengidentifikasi variabel penting, dan memahami distribusi data sebelum menerapkan metode analisis statistik yang lebih lanjut.

Dalam tugas kecil 2 ini, kami melakukan analisis data latih pada sebuah dataset untuk mengimplementasikan teknik-teknik EDA, seperti menelaah karakteristik setiap atribut dan keterkaitan antar data.

Pertama-tama, kami akan melakukan konfigurasi awal pembacaan dua buah file .csv: "data_train.csv" dan "data_validation.csv". Dalam bab analisis, kami akan membandingkan nilai-nilai dari kedua file dataset. Kami menggunakan library pandas untuk membaca file-file csv.

```python
# Import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import numpy as py
from scipy import stats
import seaborn as sns
import math
from statsmodels.stats.proportion import *

# Baca file .csv
df_train = pd.read_csv('data_train.csv')
df_validation = pd.read_csv('data_validation.csv')
```

Berikut adalah nama dan deskripsi atribut dari dataset yang akan dianalisis.

1. battery_power: Total energi baterai dalam satu waktu diukur dalam mAh
2. blue: Memiliki bluetooth atau tidak

4. dual_sim: Memiliki dukungan dual sim atau tidak
5. fc: Resolusi kamera depan dalam megapiksel
6. four_g: Memiliki 4G atau tidak
7. int_memory: Memori internal dalam gigabyte
8. m_dep: Ketebalan ponsel dalam cm
9. mobile_wt: Berat ponsel
10. n_cores: Jumlah core prosesor
11. pc: Resolusi kamera utama dalam megapiksel
12. px_height: Tinggi resolusi piksel
13. px_width: Lebar resolusi piksel
14. ram: Ukuran RAM dalam megabyte
15. sc_h: Tinggi layar ponsel dalam cm
16. sc_w: Lebar layar ponsel dalam cm
17. talk_time: Waktu telepon maksimum dalam satu kali pengisian baterai
18. three_g: Memiliki 3G atau tidak
19. touch_screen: Memiliki layar sentuh atau tidak
20. wifi: Memiliki wifi atau tidak
21. price_range (target): Rentang harga dengan nilai 0 (biaya rendah), 1 (biaya sedang), 2 (biaya tinggi) atau 3 (biaya sangat tinggi).

### **D. Outlier**

Dalam dataset ini, akan dideteksi data *outlier* yang berada di luar *range* nilai normal.

```python
# Graf boxplot data_train

fig = plt.figure(figsize=(20,15))
for i, col in enumerate(df_train.columns):
    ax = fig.add_subplot(5, 5, i+1)
    plt.boxplot(df_train[col], labels=[col])
plt.suptitle("Graf boxplot data_train")
plt.show()

```

```python
# Mencari outlier pada kolom "fc" di data_train
fc_data = df_train["fc"]
iqr = fc_data.quantile(.75) - fc_data.quantile(.25)
print("Batas atas whisker: " + str(fc_data.quantile(.75) + 1.5 * iqr))
df_train_fc_outlier = df_train.loc[(fc_data < fc_data.quantile(.25) - 1.5 * iqr) | (fc_data > fc_data.quantile(.75) + 1.5 * iqr)]
print("Banyak data outlier: " + str(len(df_train_fc_outlier["fc"])))
print(df_train_fc_outlier["fc"])
```

```python
# Graf boxplot data_validation

fig = plt.figure(figsize=(20,15))
for i, col in enumerate(df_validation.columns):
    ax = fig.add_subplot(5, 5, i+1)
    plt.boxplot(df_validation[col], labels=[col])
plt.suptitle("Graf boxplot data_validation")
plt.show()
```

```python
# Mencari outlier pada kolom "fc" di data_validation
fc_data_2 = df_validation["fc"]
iqr_2 = fc_data_2.quantile(.75) - fc_data_2.quantile(.25)
print("Batas atas whisker: " + str(fc_data_2.quantile(.75) + 1.5 * iqr_2))
df_validation_fc_outlier = df_validation.loc[(fc_data_2 < fc_data_2.quantile(.25) - 1.5 * iqr_2) | (fc_data_2 > fc_data_2.quantile(.75) + 1.5 * iqr_2)]
print("Banyak data outlier: " + str(len(df_validation_fc_outlier["fc"])))
df_validation_fc_outlier["fc"]
```

```python
# Mencari outlier pada kolom "px_height" di data_validation
px_height_data_2 = df_validation["px_height"]
iqr_3 = px_height_data_2.quantile(.75) - px_height_data_2.quantile(.25)
print("Batas atas whisker: " + str(px_height_data_2.quantile(.75) + 1.5 * iqr_3))
df_validation_px_height_outlier = df_validation.loc[(px_height_data_2 < px_height_data_2.quantile(.25) - 1.5 * iqr_3) | (px_height_data_2 > px_height_data_2.quantile(.75) + 1.5 * iqr_3)]
print("Banyak data outlier: " + str(len(df_validation_px_height_outlier["px_height"])))
df_validation_px_height_outlier["px_height"]
```

### **E. Distribution Graph**

Berikut adalah analisis distribusi dari dataset ini.

*   Untuk kolom numerik, kami memakai plot dan analisis kurtosis (dan histogram).
*   Untuk kolom non numerik, kami memakai bar chart.





**Kolom Numerik**

```python
# Graf plot histogram + kurtosis data_train
fig, axs = plt.subplots(5,3, figsize=(10, 15))

for i, (columnName, columnData) in enumerate(df_train_interval_ratio.items()):
  sns.set(style="whitegrid")
  sns.histplot(columnData, kde=True, ax=axs[i // 3, i % 3])
  axs[i // 3, i % 3].set_xlabel('Nilai')
  axs[i // 3, i % 3].set_ylabel('Frekuensi')
  axs[i // 3, i % 3].set_title(columnName + " (data_train)")
  axs[i // 3, i % 3].text(0.97, 1, s="Kurtosis: %.4f" % df_train_interval_ratio[columnName].kurt(),\
                          fontsize=9, verticalalignment='top', horizontalalignment='right', transform=axs[i // 3, i % 3].transAxes)

# Menghapus sisa axis yang tidak digunakan
for a in axs.flat[df_train_interval_ratio.columns.size:]:
     a.set_axis_off()

plt.suptitle("Graf histogram + kurtosis data_train")
plt.tight_layout()
plt.show()
```

```python
# Graf kurtosis data_train
fig, axs = plt.subplots(5,3, figsize=(10, 15))

for i, (columnName, columnData) in enumerate(df_train_interval_ratio.items()):
  sns.set(style="whitegrid")
  sns.kdeplot(columnData, ax=axs[i // 3, i % 3])
  axs[i // 3, i % 3].set_xlabel('Nilai')
  axs[i // 3, i % 3].set_ylabel('Densitas')
  axs[i // 3, i % 3].set_title(columnName + " (data_train)")
  axs[i // 3, i % 3].text(0.97, 1, s="Kurtosis: %.4f" % df_train_interval_ratio[columnName].kurt(),\
                          fontsize=9, verticalalignment='top', horizontalalignment='right', transform=axs[i // 3, i % 3].transAxes)

# Menghapus sisa axis yang tidak digunakan
for a in axs.flat[df_train_interval_ratio.columns.size:]:
     a.set_axis_off()

plt.suptitle("Graf kurtosis data_train")
plt.tight_layout()
plt.show()
```

```python
# Graf plot histogram + kurtosis data_validation
fig, axs = plt.subplots(5,3, figsize=(10, 15))

for i, (columnName, columnData) in enumerate(df_validation_interval_ratio.items()):
  sns.set(style="whitegrid")
  sns.histplot(columnData, kde=True, ax=axs[i // 3, i % 3])
  axs[i // 3, i % 3].set_xlabel('Nilai')
  axs[i // 3, i % 3].set_ylabel('Frekuensi')
  axs[i // 3, i % 3].set_title(columnName + " (data_validation)")
  axs[i // 3, i % 3].text(0.97, 1, s="Kurtosis: %.4f" % df_validation_interval_ratio[columnName].kurt(),\
                          fontsize=9, verticalalignment='top', horizontalalignment='right', transform=axs[i // 3, i % 3].transAxes)

# Menghapus sisa axis yang tidak digunakan
for a in axs.flat[df_validation_interval_ratio.columns.size:]:
     a.set_axis_off()

plt.suptitle("Graf histogram + kurtosis data_validation")
plt.tight_layout()
plt.show()
```
