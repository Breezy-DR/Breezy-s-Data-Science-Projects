## Read dataset

```python
df = pd.read_csv('/kaggle/input/wellbeing-and-performance-dataset/DataScience_QBitness_StudentWellbeingandPerformanceDashboard.csv')
```

```python
df.head()
```

## Remove duplicates

```python
df.drop_duplicates(subset='student_id', inplace=True)
```

## Handle missing values

```python
missing_values = df.isnull().sum()

missing_values = missing_values[missing_values > 0]
if missing_values.empty:
    print("There are no missing values in any of the columns")
else:
    print("Columns with missing values:\n", missing_values)
```

## Feature engineering

```python
# Mental Health Cetegory
def categorize_mental_health(level):
    if level <= 3:
        return 'Low'
    elif level <= 6:
        return 'Medium'
    else:
        return 'High'

df['mental_health_category'] = df['mental_health_rating'].apply(categorize_mental_health)

# Stress Category
def categorize_stress(level):
    if level <= 3:
        return 'Low'
    elif level <= 6:
        return 'Medium'
    else:
        return 'High'

df['stress_category'] = df['stress_level'].apply(categorize_stress)

# Exam Anxiety Category
def categorize_exam_anxiety(level):
    if level <= 3:
        return 'Low'
    elif level <= 6:
        return 'Medium'
    else:
        return 'High'

df['exam_anxiety_category'] = df['exam_anxiety_score'].apply(categorize_exam_anxiety)

# Exercise Frequency Category
def categorize_exercise(level):
    if level <= 1:
        return '0-1 Time'
    elif level <= 3:
        return '2-3 Times'
    elif level <= 5:
        return '4-5 Times'
    else:
        return '6-7 Times'

df['exercise_frequency_category'] = df['exercise_frequency'].apply(categorize_exercise)
```

## Save cleaned dataset
